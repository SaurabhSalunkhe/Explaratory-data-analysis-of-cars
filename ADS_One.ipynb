```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15, 9)
sns.set_style('whitegrid')
import numpy as np
%matplotlib inline
```

```python
# reading the data set
car_data=pd.read_csv('car_ad.csv',encoding="ISO-8859-1")
```

```python
# taking a back-up of our data set
# saving the original data
car_data_backup=car_data.copy()
car_data
```

```python
# Identifying the missing values
car_data.isnull().sum()
# Here, we can see that there are 434 missing values in engV and 511 missing values in drive
```

```python
# Here, as a placeholder, I am simply using the string "missing" for the column drive
car_data['drive'].fillna('missing',inplace=True)
```

```python
# There are multiple ways for imputing missing data like a constant value, mean, median, mode or a
# value estimated from another predicted column
# Here, I am using the mode as a placeholder for null values for engV column
car_data['engV'].mode()
# imputing the data here with the mode value
car_data['engV'].fillna(2,inplace=True)
```

```python
# Lets plot histogram for price using the plot function (OOP style)
car_data['price'].plot(kind='hist')
plt.xlabel('Price')
# Summary : The data is highly imbalanced. As observed the histogram in not normally distributed. It is right skewed histogram
```

```python
car_data['mileage'].plot(kind='hist')
plt.xlabel('Mileage')
# Summary : Again, here we can see that the data is right skewed 
```

```python
car_data['engV'].plot(kind='hist')
plt.xlabel('engV')
# Summary: The engV column is also right skewed with dominant values.
```

```python
# The pandas describe method proves to be very useful while giving the summary statistics of the data-sets
car_data.describe()
# From the summary statistics, we can conculde by looking at the min value of price and mileage that, the data may contain some 
# in appropriate values since price of a car or mileage of a car cannnot be 0. 
```

```python
# Scaling of the variables:
# Standardization of data-sets is a common requirement for many machine learning estimators
# They might behave badly if individual data column does not seem to be properly normally distributed
# In my case, the data-sets, as seen from the above graphs contains a lot of outliers, so it is essential to scale some of the
# variables
```

```python
from sklearn import preprocessing
car_price=car_data['price']
car_price_scaled = preprocessing.scale(car_price)
car_price_scaled_df=pd.DataFrame({'car_price_scaled':car_price_scaled.tolist()}) 
sns.distplot(car_price_scaled_df)
```
